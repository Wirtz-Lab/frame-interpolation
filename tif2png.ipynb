```python
import os
import glob
from PIL import Image
import cv2
import numpy as np
from matplotlib import pyplot as plt
from natsort import natsorted
src = r'\\shelter\Pratik\KYU\Raw Images\GT22\8HOUR\REP2'
ims = glob.glob(os.path.join(src,'*tif'))
ims = natsorted(natsorted)
```

```python
dst1 = r'\\shelter\Pratik\KYU\Raw Images\GT22\8HOUR\REP2\nuc50'
dst2 = r'\\shelter\Pratik\KYU\Raw Images\GT22\8HOUR\REP2\act50'
if not os.path.exists(dst1): os.mkdir(dst1)
if not os.path.exists(dst2): os.mkdir(dst2)
ims_tmpc1 = [_ for _ in ims if "ch00" in _]
ims_tmpc2 = [_ for _ in ims if "ch01" in _]
for im in ims_tmpc1:
    imnm = im.replace('tif','png')
    fol,imnm = os.path.split(imnm)
    im = Image.open(im)
    imsz = im.size()
    im = np.array(im)

    scale_percent = 50 # percent of original size
    width = round(im.shape[1] * scale_percent / 100)
    height = round(im.shape[0] * scale_percent / 100)
    dim = (width, height)
    # resize image
    im = cv2.resize(im, dim, interpolation = cv2.INTER_AREA)

    im = cv2.normalize(im,None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    Image.fromarray(im).save(os.path.join(dst1,imnm))
for im in ims_tmpc2:
    imnm = im.replace('tif','png')
    fol,imnm = os.path.split(imnm)
    im = Image.open(im)
    im = np.array(im)

    scale_percent = 50 # percent of original size
    width = round(im.shape[1] * scale_percent / 100)
    height = round(im.shape[0] * scale_percent / 100)
    dim = (width, height)
    # resize image
    im = cv2.resize(im, dim, interpolation = cv2.INTER_AREA)

    im = cv2.normalize(im,None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    Image.fromarray(im).save(os.path.join(dst2,imnm))
```

```python
import tqdm
from natsort import natsorted
import os
from PIL import Image
src = r'\\shelter\Pratik\KYU\Raw Images\GT22\8HOUR\REP2\act'
dst = os.path.join(src,'resized')
if not os.path.exists(dst): os.mkdir(dst)
ims = [_ for _ in os.listdir(src) if _.endswith('.png')]
ims = natsorted(ims)
ims
```

```python
for im in tqdm.tqdm(ims):
    resized = Image.open(os.path.join(src,im))
    imsz = resized.size
    resized = resized.resize((round(imsz[0]/2),round(imsz[1]/2)))
    resized.save(os.path.join(dst,im))
```
